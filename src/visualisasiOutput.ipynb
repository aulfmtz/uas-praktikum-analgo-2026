```python
import matplotlib.pyplot as plt
import numpy as np

algoritma = ['GREEDY\n(Heuristik)', 'DFS PRUNING\n(Eksak)']
waktu_ms = [0.0253, 6.1015]
tco_subsidi = [4992.17, 5295.98]
tco_krisis = [19964.90, 20268.71]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

bars1 = ax1.bar(algoritma, waktu_ms, color=['#2ca02c', '#d62728'])
ax1.set_title('Komparasi Waktu Eksekusi Algoritma', fontsize=14, pad=15)
ax1.set_ylabel('Milidetik (ms)', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height} ms',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),
                 textcoords="offset points",
                 ha='center', va='bottom', fontweight='bold')

x = np.arange(len(algoritma))
width = 0.35 

bars_sub = ax2.bar(x - width/2, tco_subsidi, width, label='Skenario Subsidi', color='#1f77b4')
bars_kris = ax2.bar(x + width/2, tco_krisis, width, label='Skenario Krisis', color='#ff7f0e')

ax2.set_title('Total Cost of Ownership (TCO)', fontsize=14, pad=15)
ax2.set_ylabel('Rupiah (Rp)', fontsize=12)
ax2.set_xticks(x)
ax2.set_xticklabels(algoritma)
ax2.legend()
ax2.grid(axis='y', linestyle='--', alpha=0.7)

def labelling(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'Rp {height:,.0f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

labelling(bars_sub, ax2)
labelling(bars_kris, ax2)

plt.tight_layout()
plt.savefig('komparasi_algoritma.png', dpi=300)
plt.show()
```
